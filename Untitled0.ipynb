```python
from diffusers import StableDiffusionPipeline, UNet2DConditionModel
import torch
model_path = "/content/diffusers/examples/text_to_image/sd-pokemon-model/checkpoint-15000"

unet = UNet2DConditionModel.from_pretrained(model_path + "/unet")
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", unet=unet)
pipe.to("cuda")

image = pipe(prompt="a drawing of a green pokemon with red eyes").images[0]
image.save("pokemon.png")
```

```python
import torch
from diffusers import StableDiffusionPipeline

model_path = "/content/diffusers/examples/text_to_image/sd-pokemon-model"
pipe = StableDiffusionPipeline.from_pretrained(model_path, torch_dtype=torch.float16)
pipe.to("cuda")

image = pipe(prompt="a drawing of a green pokemon with red eyes").images[0]
image.save("fancy-pokemon.png")
```
